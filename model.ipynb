# Model workflow

Import dependencies

```python
import matplotlib.pyplot as plt
from modelpkg.construct import (
    Model, 
    simulate_salaries
)
```

Simulate 1,000 pairs of salary and age

```python
age, salary = simulate_salaries(n=1000)
```

Plot data

```python
plt.scatter(x=age, y=salary)
plt.xlabel("age")
plt.ylabel("salary (DKK)")
plt.title("Salary versus age")
plt.show()
```

It appears there is a linear relationship between salary and age.

Instantiate and train model

```python
m = Model()
m.train(age, salary) 
```

Compute predicted salaries and compare with observed outcomes

```python
# predictions
predictions = m.predict(age)
# plot
ax = plt.gca()
ax.scatter(age, salary, color="b")
ax.scatter(age, predictions, color="r")
plt.xlabel("age")
plt.ylabel("salary (DKK)")
plt.title("Actual versus predicted salaries")
plt.gca().legend(("actual", "predicted"))
plt.show()
```
